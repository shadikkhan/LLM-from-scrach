## Pretreain google model word2vec

```python
import gensim.downloader as api
model = api.load('word2vec-google-news-300')
```

```python
word_vector = model
print(word_vector['computer'])
```

```python
print(word_vector['computer'].shape)
```

```python
# Eample of keeping word sementic meaning
print(word_vector.most_similar(positive=['king', 'woman'],
                               negative=['man'],
                               topn=10))
```

```python
print(word_vector.similarity('woman', 'man'))
print(word_vector.similarity('king', 'queen'))
print(word_vector.similarity('uncle', 'aunt'))
print(word_vector.similarity('boy', 'girl'))
print(word_vector.similarity('nephew', 'niece'))
print(word_vector.similarity('paper', 'water'))
print(word_vector.similarity('cat', 'dog'))
print(word_vector.similarity('apple', 'banana'))
print(word_vector.similarity('computer', 'man'))
```
